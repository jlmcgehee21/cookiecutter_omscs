```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def lecture_coding_example(arg1, arg2):
    x = np.arange(0, 10, 0.01)
    return x, np.sin(arg1*x) + arg2

time, output = lecture_coding_example(3, 10)

figure = plt.plot(time, output)
```
